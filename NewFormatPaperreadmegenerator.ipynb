```python
def removeSpaces(string):
 
    
    count = 0
 
    list = []
 
  
    for i in range(len(string)):
        if string[i] != ' ':
            list.append(string[i])
 
    return toString(list)
 
# Utility Function
def toString(List):
    return ''.join(List)
```

```python
def stringClean(String1):
    
    val1 =re.sub(",","",String1)
    val2 =re.sub("-","",val1)
    val3 =re.sub(":","",val2)
    
    return val3
```

```python
paperlist = []
paperlist.append(t3.substitute(Papers = "Papers",
                               papers = "Papers"))
for i in range(len(val)):
    paperlist.append(t2.substitute(
                                   year = val[i]))
paperlist.append(t4.substitute(Papers = "Papers"))
```

```python
for i in range(len(val)):
    paperlist.append(t1.substitute(year = val[i]))
    df = p1.get_group(val[i])
    for ind in df.index:
        paperlist.append(t.substitute(papername=df['Title'][ind],
                      paperauthors = df['Authors'][ind],
                      paperurl = df['PaperURL'][ind],
                      websiteurl1 = df['Website1'][ind],
                      websiteurl2 = df['Website2'][ind],
                                      
                      videourl1 = df['Video1'][ind],
                      videourl2 = df['Video2'][ind],
                      videourl3 = df['Video3'][ind],
                      giturl1 = df['Github1'][ind],
                      giturl2 = df['Github2'][ind],
                      giturl3 = df['Github3'][ind],
                      
                      conf       = df['Venue'][ind],
                      newpapername = removeSpaces(stringClean(df['Title'][ind]))            
                      ))
        
```

```python
readmefile = open("readme2.md",'w',encoding="utf-8")
for i in paperlist:
    readmefile.write(i+'\n')
readmefile.close()
```
